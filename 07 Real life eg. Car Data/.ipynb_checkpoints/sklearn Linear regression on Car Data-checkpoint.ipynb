```python
# Since we will be using the log price variable, we can drop the old 'Price' one
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

#### Multicollinearity

```python
data_cleaned.columns.values
```

```python
# sklearn does not have a built-in way to check for multicollinearity
# one of the main reasons is that this is an issue well covered in statistical frameworks and not in ML ones

from statsmodels.stats.outliers_influence import variance_inflation_factor

# To make this as easy as possible to use, we declare a variable where we put
# all features where we want to check for multicollinearity
# since our categorical data is not yet preprocessed, we will only take the numerical ones
variables = data_cleaned[['Mileage','Year','EngineV']]

# we create a new data frame which will include all the VIFs (Variance Inflation factor)
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)
vif = pd.DataFrame()

# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = variables.columns
```

```python
vif
```

```python
# Since Year has the highest VIF, we will remove it from the model
# This will drive the VIF of other variables down!!! 
# So even if EngineV seems with a high VIF, too, once 'Year' is gone that will no longer be the case
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```

#### Creating Dummy Variables

```python
# Pandas provide method, which spots all categorical variables and create dummies automatically!
```

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
# drop_first=true arg will not assign any dummies for audi(first)
```

```python
data_with_dummies.head()
```

```python
# One line of code preprocessed all categorical features!!!
```

#### Rearrange the columns 

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear Regression Model

#### Declare the inputs and targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

#### Scaling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

#### Training-Test split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=42)
```

#### Regression begins

```python
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=16)
```

```python
reg.score(x_train,y_train)
# Note that this is NOT the adjusted R-squared
```

#### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
# Weights interpretration
# A +ve weight shows thats as a feature increases in value, so the log_price and price respectively
# A -ne weight shows that as a feature increases in value, the log_price and price decreases
```

```python
data_cleaned['Brand'].unique()
```

```python
# Audi is benchmark. So a +ve brand will be expensive than audi. And -ve brand will be cheaper!
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()

# alpha=0.2 turns it into a heatmap
# aplha contains value from 0 to 1
```

```python
# To obtain the actual prices, we take the exponential of the log_price
# use np.exp() method
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
```

```python
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
